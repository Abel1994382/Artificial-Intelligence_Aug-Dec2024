```
# Imprimir nombre y matrícula
print("Name: Abel Antonio Ramírez Cavazos")
print("Student ID: 1994382")
print()

# Abrir y leer el archivo
with open('texto.rtf', 'r') as file:
    content = file.read()

# Convertir todo el contenido a minúsculas
content = content.lower()

# Eliminar puntuación de otra manera
clean_content = ""
for char in content:
    if char.isalnum() or char.isspace():  # Solo mantiene letras y números, además de espacios
        clean_content += char

# Contar el número de palabras
words = clean_content.split()
num_words = len(words)

# Contar el número de párrafos
num_paragraphs = content.count('\n\n') + 1

# Contar el número de oraciones (dividiendo por '.')
sentences = [sentence for sentence in clean_content.split('.') if sentence.strip()]
num_sentences = len(sentences)

# Calcular la longitud promedio de las palabras
total_length_words = 0
for word in words:
    total_length_words += len(word)

avg_word_length = total_length_words / num_words if num_words > 0 else 0

# Contar la frecuencia de las palabras
word_counts = {}
for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# Encontrar las palabras más frecuentes
top_n = 5
sorted_words = []
for word, count in word_counts.items():
    sorted_words.append((word, count))

# Ordenar las palabras por frecuencia de mayor a menor
for i in range(len(sorted_words)):
    for j in range(i + 1, len(sorted_words)):
        if sorted_words[i][1] < sorted_words[j][1]:
            sorted_words[i], sorted_words[j] = sorted_words[j], sorted_words[i]

top_words = sorted_words[:top_n]

# Imprimir resultados
print(f"Words: {num_words}")
print(f"Paragraphs: {num_paragraphs}")
print(f"Sentences: {num_sentences}")
print(f"Average Length: {avg_word_length:.2f}")
print("Most Common Words and Frequencies:")
for word, freq in top_words:
    print(f"{word}: {freq}")
```

```
t
qfuidef choose_word():
    words = ['algorithm', 'debugging', 'syntax', 'variable', 'function']
    word_index = len('Antonio') % len(words)
    return words[word_index]

def display_word(word, guessed_letters):
    displayed_word = ''
    for letter in word:
        if letter in guessed_letters:
            displayed_word += letter
        else:
            displayed_word += '_'
    return displayed_word

def hangman():
    print("Name: Abel Antonio Ramírez Cavazos")
    print("Student ID: 1994382")
    print()

    word = choose_word()
    guessed_letters = []
    attempts = 6

    print("Welcome to Hangman!")
    print(f"You have this {attempts} attempts to guess.")
    print(display_word(word, guessed_letters))

    while attempts > 0:
        guess = input("Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Please enter 1 letter.")
            continue

        if guess in guessed_letters:
            print("You already guessed that.")
            continue

        guessed_letters.append(guess)

        if guess in word:
            print("Good guess!")
        else:
            attempts -= 1
            print(f"Incorrect letter. You have {attempts} attempts left.")

        print(display_word(word, guessed_letters))

        # Verificar si todas las letras fueron adivinadas
        won = True
        for letter in word:
            if letter not in guessed_letters:
                won = False
                break

        if won:
            print("Congratulations!")
            break
    else:
        print(f"The word was '{word}'.")

if __name__ == "__main__":
    hangman()
```
